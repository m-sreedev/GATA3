```python
#pvalue checker
namelist = ["DP","CD4","CD8","Th1","Th2","Th17","iTreg", "nTreg", "NKT", "CD8-Fli1","CD8-KO-Fli","Th2-Fli1","DP-Replicate", "CD8-KO","Th2 Replicate"]
from scipy.stats import hypergeom
import pandas as pd
def pvalues(celltype):
    data = histogrammer(celltype)
    mvalues = data[1]
    pdict = {}
    for i in range(len(mvalues)):
        capm = int(mvalues[celltype-1])
        for j in range(len(data[0].keys())):
            k = int(data[0][str(j+1)][celltype-1])
            plist = []
            for l in range(len(mvalues)):
                capn = int(mvalues[l])
                x = int(data[0][str(j+1)][l])
                #print(capm,k,capn,x)
                dist = hypergeom(capm,k,capn)
                dlist = [capm,k,capn,x]
                if x/capn >= k/capm:
                    pval = [round(dist.sf(x-1),5),"+"]
                else:
                    pval = [round(dist.cdf(x-1),5),"-"]
                plist.append(pval)
            pdict[str(j+1)] = plist
    df = pd.DataFrame(pdict)
    df.insert(0,"Cell Type",namelist)
    return df

pvalues(1)
```

```python
def htmler():
    namelist = [0,"DP","CD4","CD8","Th1","Th2","Th17","iTreg", "nTreg", "NKT", "CD8-Fli1","CD8-KO-Fli","Th2-Fli1","DP-Replicate", "CD8-KO","Th2 Replicate"]
    for i in range(15,16):
        baby = pvalues(i)
        baby.insert(0,namelist[i],namelist[1:])
        baby.to_html("/home/sreedev/Downloads/Output/Plots/P Values//"+str(i)+".html",index=False)
```

```python
def stylise():
    folderloc = "/home/sreedev/Downloads/Output/Plots/P Values//"
    setting = """    <style>
            table {
                border-collapse: collapse;
            }
            td {
                border: 1px solid black;
                padding: 10px;
            }
            img {
                max-width: 100%;
                max-height: 100%;
            }
        </style>\n"""
    for i in range(15,16):
        html = open(folderloc + str(i) + "//" + str(i) + ".html","r+")
        new_html = open(folderloc + str(i) + "//" + str(i) + "c.html","w")
        lines = html.readlines()
        motifno = -1
        for line in lines:
            if line[6:10] == "<th>":
                motifno +=1
        imagelocs = []
        for j in range(1,motifno+1):
            if j < 10:
                imageloc = folderloc + str(i) + "//" + "sites_0" + str(j) + ".png"
            else:
                imageloc = folderloc + str(i) + "//" + "sites_" + str(j) + ".png"
            location = "      <th><img src = \"" + str(imageloc) + "\" </img></th>\n"
            imagelocs.append(location)
        for k in range(4,motifno+4):
                lines.pop(k)
                lines.insert(k,imagelocs[k-4])

        lines.insert(0,setting)

        for line in lines:
            new_html.write(line)

```

```python
import pandas as pd
def weedout(celltype):
    df = pvalues(celltype)
    pdict = df.to_dict(orient = 'dict')
    for key in pdict.keys():
        valuelist = pdict[key]
        for i in range(0,len(valuelist)):
            if valuelist[i] < 0.05:
                print(key, i)
```

```python
def highlighter(x):
    condition = x[0] < 0.05
    bg_color = 'background-color : #242020' if condition else ''
    return bg_color

def weedf(celltype):
    import numpy as np
    df0 = pvalues(celltype)
    df = df0.iloc[:,np.arange(1,len(df0.columns))]
    styled = df.style.applymap(highlighter)
    return styled

weedf(1)
```
